```python
N = train["Resolution"].size
second_res = []
for i in xrange(0,N):
    sp = train["Resolution"][i].split(", ")
    if len(sp) > 1:
        second_res.append(sp[1])
    else:
        second_res.append(sp[0])
train["ResolutionSecond"]=second_res
```

```python

first_res = []
for i in xrange(0,N):
    first_res.append(train["Resolution"][i].split(", ")[0])
train["Resolution"]=first_res

train["Dates"]=pd.to_datetime(train["Dates"])
test["Dates"]=pd.to_datetime(test["Dates"])
isNight = []
hours = []
months = []
years = []
for i in xrange(0,N):
    year = train["Dates"][i].year
    month = train["Dates"][i].month
    hour = train["Dates"][i].hour
    if hour >= 22 or (hour >= 0 and hour < 6) : isNight.append(1)
    else : isNight.append(0)
    hours.append(hour)
    years.append(year)
    months.append(month)

train["isNight"]=isNight
train["Hours"]=hours
train["Months"]=months
train["Years"]=years

M=test["Dates"].size
isNight = []
hours = []
months = []
years = []
for i in xrange(0,M):
    year = test["Dates"][i].year
    month = test["Dates"][i].month
    hour = test["Dates"][i].hour
    if hour >= 22 or (hour >= 0 and hour < 6) : isNight.append(1)
    else : isNight.append(0)
    hours.append(hour)
    years.append(year)
    months.append(month)

test["Hours"]=hours
test["isNight"]=isNight
test["Months"]=months
test["Years"]=years
```
