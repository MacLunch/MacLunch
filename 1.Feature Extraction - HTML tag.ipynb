```python
from html.parser import HTMLParser as hp
import html
import re
import copy
import numpy as np
import pandas as pd
import json
import glob
from random import shuffle
from bs4 import BeautifulSoup as bs
```

### file read

```python
data_path = "data"
#data_path = "data(test)"

file_list = glob.glob("%s/*.json" % data_path)

json_train=[]

shuffle(file_list)
for json_file_name in file_list:
    print(json_file_name)
    json_file = json.loads(open(json_file_name).read())
    json_train += json_file["articles"]
    
print(len(json_train))
```

### get tag type & make DF

```python
tag_list = []
label_list = []

num_json_train = len(json_train)

for i, article in enumerate(json_train):
    if i % 10000 == 0:
        print("converting... (%d/%d)" % (i, num_json_train))

    tags = {}

#     train_text = article["text"]
#     matches = re.findall('<[^/].*?>', train_text)
    
#     for tag in matches:
#         tag_name = 'TAG_' + tag.replace('<','').replace('>','').strip().split(' ')[0]
#         #print(tag_name)
#         if tag_name not in tags:
#             tags[tag_name] = 0
#         tags[tag_name] += 1
    
    train_text = article["text"]
    soup = bs(train_text, 'html.parser')
    for tag in soup.find_all():
        tag_name = 'TAG_' + tag.name
        if tag_name not in tags:
            tags[tag_name] = 0
        tags[tag_name] += 1

    tags["is_troll"] = article["is_troll"]
    tag_list.append(tags)
    

data = pd.DataFrame.from_dict(tag_list)
data = data.fillna(0)

data
```

### 1번만 사용되는 tag 제거

```python
drop_col_list = []
ex = []
for colname in data.columns.drop([label_name]):
#    print(colname, sum(data[colname]))
    if sum(data[colname]) == 1:
        drop_col_list.append(colname)
    else:
        ex.append(colname)
        
len(drop_col_list)
ex
```

```python
data = data.drop(drop_col_list, 1)
```

### Random Forest

```python
from sklearn.ensemble import RandomForestClassifier as RF

label_name = "is_troll"
predictors = data.columns.drop([label_name])

model = RF(n_estimators=10, n_jobs=4)
```

### Cross Validation

```python
from sklearn import cross_validation

cv_value = 3

scores = cross_validation.cross_val_score(model, data[predictors], data[label_name], cv=cv_value, scoring='roc_auc')
cv_result = scores.mean()

print(cv_result)
```
