### json 불러와서 저장하기

```python
import numpy as np
import pandas as pd
import json
import jpype
import glob
from random import shuffle
from bs4 import BeautifulSoup as bs

data_path = "data"
#data_path = "data(test)"

file_list = glob.glob("%s/*.json" % data_path)

json_train=[]

shuffle(file_list)
for json_file_name in file_list:
    json_file = json.loads(open(json_file_name).read())
    json_train += json_file["articles"]
```

### json에서 author, author 형태소, forumid 추출

```python
from konlpy.tag import Mecab

mecab = Mecab()

labeled_train = []

for cnt, article in enumerate(json_train):
    text = bs(article["text"],"lxml").text
    if cnt % 10000 == 0:
        print cnt

    labeled_train.append({
            "istroll": article["is_troll"],
            "author": article["author"],
            "author_pos": ["%s_%s" % (word, pos) for word, pos in mecab.pos(article["author"])],
            "text_pos": ["%s_%s" % (first, second) for first, second in mecab.pos(text)],
            "title_pos": ["%s_%s" % (word, pos) for word, pos in mecab.pos(article["title"])],
            "forumid": article["forumid"],
            "pk": article["pk"]
        })

labeled_train = pd.DataFrame.from_dict(labeled_train)
labeled_train = labeled_train.set_index('pk')

labeled_train
```

### forumid와 author을 label encoder로 숫자로 바꿔줌

```python
from sklearn import preprocessing

le = preprocessing.LabelEncoder()

labeled_train["forumid"] = le.fit_transform(labeled_train["forumid"])
labeled_train["author"] = le.fit_transform(labeled_train["author"])

labeled_train
```

##Word2Vec

```python
from gensim.models import word2vec

def makeFeatureVec(words, model, num_features):
    
    featureVec = np.zeros((num_features,), dtype = "float32")
    
    nwords = 0
    
    index2word_set = set(model.index2word)
    
    for word in words:
        if word in index2word_set:
            nwords = nwords + 1.
            featureVec = np.add(featureVec, model[word])
    
    if nwords != 0:
        featureVec = np.divide(featureVec, nwords)
    
    return featureVec

def getAvgFeatureVecs(texts, model, num_features):
    
    counter = 0
    
    textFeatureVecs = np.zeros((len(texts), num_features), dtype = "float32")
    
    for i, text in enumerate(texts):
        
        if i % 10000 == 0:
            print(i)
        
        textFeatureVecs[i] = makeFeatureVec(text , model, num_features)
        
    return textFeatureVecs

def make_word2vec_v1(train):
    num_features = 1000
    min_word_count = 40
    num_workers = 4
    context = 10
    downsampling = 1e-3
    
    combined_train = train["author_pos"].apply(lambda x:" ".join(x))
    combined_train += train["title_pos"].apply(lambda x:" ".join(x))
    combined_train += train["text_pos"].apply(lambda x:" ".join(x))
    
    sentences = " ".join(combined_train)
    
    model = word2vec.Word2Vec(sentences, workers = num_workers, size = num_features,\
                             min_count = min_word_count, window = context, sample = downsampling)
    
    col = ["word2vec_%s_%d" % ("all", data) for data in range(0, num_features)]
    
    train_feature = getAvgFeatureVecs(combined_train, model, num_features)
    train_feature = pd.DataFrame(train_feature, index = train.index, columns = col)
    
    train = pd.concat([train, train_feature], axis = 1)
    
    return train

def make_word2vec_v2(train):
    num_features = 1000
    min_word_count = 40
    num_workers = 4
    context = 10
    downsampling = 1e-3
    
    combined_train1 = train["author_pos"].apply(lambda x:" ".join(x))
    combined_train2 = train["title_pos"].apply(lambda x:" ".join(x))
    combined_train3 = train["text_pos"].apply(lambda x:" ".join(x))
    
    sentences = " ".join(combined_train1 + combined_train2 + combined_train3)
    
    combined_train = train["author_pos"].apply(lambda x:" ".join(x))
    combined_train += train["title_pos"].apply(lambda x:" ".join(x))
    combined_train += train["text_pos"].apply(lambda x:" ".join(x))
    
    model = word2vec.Word2Vec(sentences, workers = num_workers, size = num_features,\
                             min_count = min_word_count, window = context, sample = downsampling)
    
    col = ["word2vec_%s_%d" % ("all", data) for data in range(0, num_features)]
    
    train_feature = getAvgFeatureVecs(combined_train, model, num_features)
    train_feature = pd.DataFrame(train_feature, index = train.index, columns = col)
    
    train = pd.concat([train, train_feature], axis = 1)
    
    return train
```

```python
%time train = make_word2vec_v1(labeled_train)
```

```python
import xgboost as xgb
from sklearn.ensemble import RandomForestClassifier

label = 'istroll'
pre = train.columns.drop([label, "forumid", "author", "author_pos", "text_pos", "title_pos"])

model = RandomForestClassifier(n_estimators=10, n_jobs=6)
```

```python
from sklearn import cross_validation

cv_value = 3

scores = cross_validation.cross_val_score(model, train[pre], train[label], cv=cv_value, scoring="roc_auc")
cv_result = scores.mean()

submission = pd.DataFrame()

cv_value = str(cv_result)
submission.to_csv("word2vec v1 - %s.csv" % cv_value)
```
